# Get some venues for Washington DC off the API

```python
# Get first 300 venues from Washington DC with 3 seperate api calls (offset parameter goes from 0 - 60)
LIMIT = 100
radius = 1500

url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&offset={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    neighborhood_latitude, 
    neighborhood_longitude, 
    radius, 
    LIMIT,
    0)
```

```python
#Get requests based on the above urls
results = requests.get(url).json()
results['response']['groups'][0]['items'][0]['venue']['location']['neighborhood']
```

```python
venueName = []
venueLat = []
venueLong = []
venueZip = []
venue_Neighborhood = []
venueCat = []



for i in range(len(results['response']['groups'][0]['items'])):
    venueName.append(results['response']['groups'][0]['items'][i]['venue']['name'])
    venueLat.append(results['response']['groups'][0]['items'][i]['venue']['location']['lat'])
    venueLong.append(results['response']['groups'][0]['items'][i]['venue']['location']['lng'])
    #venueZip.append(jsonData1['response']['groups'][0]['items'][i]['venue']['location']['postalCode'])
    venueCat.append(results['response']['groups'][0]['items'][i]['venue']['categories'][0]['name'])

dict = {'Venue_Name':venueName,'Venue_Category':venueCat, 'Venue_Lat':venueLat, 'Venue_Lng': venueLong}
data = pd.DataFrame(dict)
len(data)
```

```python
# one hot encoding
data_onehot = pd.get_dummies(data[['Venue_Category']], prefix="", prefix_sep="")
data_onehot.head()
```

```python
# set number of clusters
kclusters = 5

data_clustering = data_onehot

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(data_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
kmeans.labels_
```

# Plot points on a map

```python
import numpy as np
map_clusters = folium.Map(location=[neighborhood_latitude, neighborhood_longitude], zoom_start=16)
x=np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]


# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(data['Venue_Lat'], data['Venue_Lng'], data['Venue_Name'], kmeans.labels_):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
